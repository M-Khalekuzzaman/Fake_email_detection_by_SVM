```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('spam.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['Category'] = [1 if x == 'spam' else 0 for x in df['Category']]
```

```python
df.head()
```

```python
import nltk
```

```python
from nltk.corpus import stopwords
```

```python
from nltk.tokenize import word_tokenize
```

```python
print(stopwords)
```

```python
nltk.download('stopwords')
```

```python
nltk.download('punkt')
```

```python
stop_words = set(stopwords.words('english'))
```

```python
stop_words
```

```python
def preprocess_text(text):
    words = word_tokenize(text)
    filtered_words = [word for word in words if word.casefold() not in stop_words]
    return ' '.join(filtered_words)
```

```python
df.head()
```

```python
x = df['Message']
y = df['Category']
```

```python
x
```

```python
x = x.apply(preprocess_text)
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer()
```

```python
x_train = tfidf.fit_transform(xtrain)
x_test = tfidf.transform(xtest)
```

```python
x_train.toarray()
```

```python
#Using ML Algorithm
from sklearn.naive_bayes import MultinomialNB
```

```python
model = MultinomialNB()
```

```python
model.fit(x_train,ytrain)
```

```python
model.score(x_test,ytest)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python

from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
print(accuracy_score(y_pred,ytest))
```

```python
print(confusion_matrix(y_pred,ytest))
```

```python
# SVM = Support Vector Machine
from sklearn.svm import SVC
```

```python
svc_model = SVC()
```

```python
svc_model.fit(x_train, ytrain)
```

```python
svc_model.score(x_test,ytest)
```

```python
ypred = svc_model.predict(x_test)
```

```python
print(accuracy_score(ypred,ytest))
```
